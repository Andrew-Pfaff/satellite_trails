```python
PNG_DIR = Path('/content/drive/MyDrive/satellite_trails/pngs')
MASTER_SPLIT_CSV = Path('/content/drive/MyDrive/satellite_trails/master_split.csv')
UNET_CHECKPOINT = Path('/content/drive/MyDrive/satellite_trails/unet_colab_shift_trial25_long_weights.pt')
CLASSIFIER_CHECKPOINT = Path('/content/drive/MyDrive/satellite_trails/classifier_weights.pt')
OUTPUT_DIR = Path('/content/drive/MyDrive/satellite_trails/final_eval_outputs')

SPLIT_ID = 2
UNET_THRESHOLD = 0.6
CLASSIFIER_THRESHOLD = 0.67
NORMALIZATION = 'source_zscore'
PATCH_DIM = 528
NUM_WORKERS = 2
```

```python
!python scripts/python/evaluate_final_full_field.py \
  --png-dir "{PNG_DIR}" \
  --master-split-csv "{MASTER_SPLIT_CSV}" \
  --unet-checkpoint "{UNET_CHECKPOINT}" \
  --classifier-checkpoint "{CLASSIFIER_CHECKPOINT}" \
  --split-id "{SPLIT_ID}" \
  --unet-threshold "{UNET_THRESHOLD}" \
  --classifier-threshold "{CLASSIFIER_THRESHOLD}" \
  --normalization "{NORMALIZATION}" \
  --patch-dim "{PATCH_DIM}" \
  --num-workers "{NUM_WORKERS}" \
  --output-dir "{OUTPUT_DIR}"
```

```python
import pandas as pd

aggregate = pd.read_csv(OUTPUT_DIR / 'aggregate_metrics.csv')
aggregate.sort_values('iou', ascending=False)
```
